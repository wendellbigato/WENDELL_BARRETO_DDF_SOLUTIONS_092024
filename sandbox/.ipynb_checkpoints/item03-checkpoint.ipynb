```python
# Define the schema of the expected JSON output
response_schemas = [
    ResponseSchema(name="category", description="The category of the product."),
    ResponseSchema(name="material", description="The material of the product."),
    ResponseSchema(name="functions", description="The functions of the product in JSON format with keys like card_slots, kickstand, etc."),
    ResponseSchema(name="compatibility", description="The compatibility of the product with other items, if applicable."),
    ResponseSchema(name="other_details", description="Other relevant details of the product.")
]

# Create a parser with LangChain to validate the output format
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)

# Função para extrair features usando um LLM e garantir o esquema correto com LangChain
def extrair_features_com_parser(titulo, descricao):
    prompt_template = """
    Titulo: {titulo}
    Descrição do Produto: {descricao}
    
    Quais são as características do produto? Crie uma lista de características em formato JSON, incluindo:
    - Categoria do produto
    - Material
    - Funções (ex: slots para cartão, espelho cosmético, suporte)
    - Compatibilidade
    - Outros detalhes relevantes.
    
    O formato do JSON deve ser o seguinte:
    {{
        "category": "...",
        "material": "...",
        "functions": {{
            "card_slots": "...",
            "cosmetic_mirror": "...",
            "kickstand_function": "..."
        }},
        "compatibility": "...",
        "other_details": "..."
    }}
    """

    # Substituir as variáveis de entrada
    prompt = prompt_template.format(titulo=titulo, descricao=descricao)

    # Chamar o LLM usando a nova API ChatCompletion
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are an assistant that extracts features from product descriptions."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=500
    )

    # Usar o parser do LangChain para validar o formato do JSON
    try:
        parsed_response = output_parser.parse(response.choices[0].message['content'].strip())
        return parsed_response
    except Exception as e:
        print(f"Erro ao validar o formato JSON: {e}")
        return None

# Carregar o CSV "linhas_validas.csv"
df = pd.read_csv("linhas_validas.csv")

# Aplicar a extração de features com o parser para garantir o formato correto
df['features'] = df.apply(lambda row: extrair_features_com_parser(row['title'], row['text']), axis=1)

# Reorganizar as colunas para que 'features' seja inserida logo após 'title'
cols = df.columns.tolist()
cols.insert(cols.index('title') + 1, cols.pop(cols.index('features')))
df = df[cols]

# Salvar o DataFrame atualizado em um novo arquivo CSV
output_csv = "linhas_validas_com_features.csv"
df.to_csv(output_csv, index=False)

print(f"Arquivo CSV '{output_csv}' salvo com sucesso com a coluna de features adicionada.")
```
